<h3>Project Overview</h3>
<p>Using what youâ€™ve learned; download the NYC Property Sales Dataset from Kaggle. This dataset is a record of every building or building unit (apartment, etc.) sold in the New York City property market over a 12-month period.</p>
<p>This dataset contains the location, address, type, sale price, and sale date of building units sold. A reference on the trickier fields:</p>
<ul><li><code>BOROUGH</code>: A digit code for the borough the property is located in; in order these are Manhattan (1), Bronx (2), Brooklyn (3), Queens (4), and Staten Island (5).</li><li><code>BLOCK</code>; <code>LOT</code>: The combination of borough, block, and lot forms a unique key for property in New York City. Commonly called a BBL.</li><li><code>BUILDING CLASS AT PRESENT</code> and <code>BUILDING CLASS AT TIME OF SALE</code>: The type of building at various points in time.</li></ul>
<p>Note that because this is a financial transaction dataset, there are some points that need to be kept in mind:</p>
<ul><li>Many sales occur with a nonsensically small dollar amount: $0 most commonly. These sales are actually transfers of deeds between parties: for example, parents transferring ownership to their home to a child after moving out for retirement.</li><li>This dataset uses the financial definition of a building/building unit, for tax purposes. In case a single entity owns the building in question, a sale covers the value of the entire building. In case a building is owned piecemeal by its residents (a condominium), a sale refers to a single apartment (or group of apartments) owned by some individual.</li></ul>

```python
df = pd.read_csv('dataset/nyc-rolling-sales.csv')
```

```python
# Data types and column names
df.info()
```

```python
# Drop unused columns
df = df.drop(columns=['Unnamed: 0', 'EASE-MENT', 'ADDRESS'])
df.info()
df.describe()
```

```python
# Drop properties with Year Built = 0
df = df[~df['YEAR BUILT'].isin([0])]
```

```python
# Convert Sale Date to date format
df['SALE_DATE'] = pd.to_datetime(df['SALE DATE'])
df['SALE_YEAR'] = df['SALE_DATE'].dt.year
```

```python
# Convert Land Square Feet, Gross Square Feet and Sale Price from object to int64
df = df.replace(' -  ', 0)
df[['LAND SQUARE FEET', 'GROSS SQUARE FEET', 'SALE PRICE']] = df[['LAND SQUARE FEET', 'GROSS SQUARE FEET', 'SALE PRICE']].astype(str).astype(int)
```

```python
# Data description
df.describe()
```

```python
# Split inheritance with true sales

df_inherit = df[df['SALE PRICE'].isin([0])]
df_true_sales = df[~df['SALE PRICE'].isin([0])]
```

```python
# Data samples - head (True Sales)
df_true_sales.head(5)
```

<h3>Chapter 2: Data Statistics</h3>

<h4>Exhibit 0 - Hypothesis Testing (Hypothesis)</h4>

<p>The following are hypothesis that will be tested on this report:
<ol>
    <li>
        Inherited properties are older than true sales properties
        <ul>
            <li>Will be analised using Mean, Median and Mode</li>
        </ul>
    </li>
    <li>
        Due to technological development to build properties, buildings built in the last 5 years (2012-2017) have more number of levels than those built 50 years ago (1962 - 1967)
        <ul>
            <li>Will be analised using Range, Variance and Standard Deviation</li>
            <li>A new parameter of ['Gross Square Area / Land Square Area'] will be calculated as it indicates the number of levels of the properties</li>
        </ul>
    </li>
    <li>
        Sales price of properties sold are very variative
        <ul>
            <li>Will be analised using Probability Distribution and Confidence Intervals</li>
        </ul>
    </li>
</ol>
</p>

<h4>Exhibit 1 - Measure of Central Tendency: Mean</h4>

```python
# Get mean value
mean_inherit_yearbuilt = statistics.mean(df_inherit['YEAR BUILT'])
mean_true_sales_yearbuilt = statistics.mean(df_true_sales['YEAR BUILT'])

# Output
print('Year Built Mean of Inherited Properties:', math.trunc(mean_inherit_yearbuilt))
print('Year Built Mean of True Sales Properties:', math.trunc(mean_true_sales_yearbuilt))
```

<h4>Exhibit 2 - Measure of Central Tendency: Median</h4>

```python
# Get median value
median_inherit_yearbuilt = statistics.median(df_inherit['YEAR BUILT'])
median_true_sales_yearbuilt = statistics.median(df_true_sales['YEAR BUILT'])

# Output
print('Year Built Median of Inherited Properties:', math.trunc(median_inherit_yearbuilt))
print('Year Built Median of True Sales Properties:', math.trunc(median_true_sales_yearbuilt))
```

<h4>Exhibit 3 - Measure of Central Tendency: Modus</h4>

```python
# Get mode value
mode_inherit_yearbuilt = statistics.mode(df_inherit['YEAR BUILT'])
mode_true_sales_yearbuilt = statistics.mode(df_true_sales['YEAR BUILT'])

# Output
print('Year Built Mode of Inherited Properties:', math.trunc(mode_inherit_yearbuilt))
print('Year Built Mode of True Sales Properties:', math.trunc(mode_true_sales_yearbuilt))
```

<h4>Exhibit 4 - Measure of Spread: Range</h4>

```python
# Calculate level parameter ['Gross Square Area / Land Square Area'] of each property and remove properties with 0 areas
df_area_clean = df[~df[~df['LAND SQUARE FEET'].isin([0])].isin([0])]
df_area_clean['LEVELS'] = np.ceil(df_area_clean['GROSS SQUARE FEET']/df_area_clean['LAND SQUARE FEET'])

# Prepare datasets for analysis
df_year_range_1 = df_area_clean[df_area_clean['YEAR BUILT'].isin(range(2012,2018))]
df_year_range_2 = df_area_clean[df_area_clean['YEAR BUILT'].isin(range(1962,1968))]
```

```python
# Get range value
range_year_range_1 = df_year_range_1['LEVELS'].max() - df_year_range_1['LEVELS'].min()
range_year_range_2 = df_year_range_2['LEVELS'].max() - df_year_range_2['LEVELS'].min()

# Output
print('Range of levels in properties built in 2012 to 2017: ', math.trunc(range_year_range_1))
print('Range of levels in properties built in 1962 to 1967: ', math.trunc(range_year_range_2))
```

<h4>Exhibit 5 - Measure of Spread: Variance</h4>

```python
# Get variance value
variance_year_range_1 = np.var(df_year_range_1['LEVELS'], ddof=1)
variance_year_range_2 = np.var(df_year_range_2['LEVELS'], ddof=1)

# Output
print('Variance of levels in properties built in 2012 to 2017: ', variance_year_range_1)
print('Variance of levels in properties built in 1962 to 1967: ', variance_year_range_2)
```

<h4>Exhibit 6 - Measure of Spread: Standard Deviation</h4>

```python
# Get standard deviation value
stdev_year_range_1 = np.std(df_year_range_1['LEVELS'], ddof=1)
stdev_year_range_2 = np.std(df_year_range_2['LEVELS'], ddof=1)

# Output
print('Standard Deviation of levels in properties built in 2012 to 2017: ', stdev_year_range_1)
print('Standard Deviation of levels in properties built in 1962 to 1967: ', stdev_year_range_2)
```

<h4>Exhibit 7 - Probability Distribution</h4>

```python
# Create probability distribution graph
ax = sns.distplot(df_true_sales['SALE PRICE'],
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15, 'alpha': 1})
ax.set(xlabel='Distribution', ylabel='Frequency', title='Year Built')
```

<h4>Exhibit 8 - Confidence Intervals</h4>

```python
# Describe the sale price data of true sales group

df_true_sales['SALE PRICE'].describe().astype(int)
```

```python
# Calculate confidence interval of 95%

# CI components
z_score = 1.96 # 95% confidence level
mean_price = np.mean(df_true_sales['SALE PRICE'])
stdev_price = np.std(df_true_sales['SALE PRICE'])
size_price = np.size(df_true_sales['SALE PRICE'])

# Calculate boundary
lcb = mean_price - z_score * stdev_price / np.sqrt(size_price)
ucb = mean_price + z_score * stdev_price / np.sqrt(size_price)

# Output
print('Upper Boundary: ', ucb)
print('Lower Boundary: ', lcb)
```
